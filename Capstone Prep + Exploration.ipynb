```python
import time
for index, app in enumerate(appid_list[375:]):
    parameters = {'filter':'all', 'langauge':'en', 'day_range' : 365,  'cursor' : '*', 
                      'review_type' : 'all', 'purchase_type' : 'all', 'num_per_page' : 100}
    for i in range(10):
        temp_url = "https://store.steampowered.com/appreviews/{appid}?json=1".format(appid = app)
        response = requests.get(temp_url, params = parameters)
        json = response.json()
        temp_df = pd.DataFrame.from_records(json['reviews'])
        steam_reviews = pd.concat([steam_reviews, temp_df])
        if len(json['reviews']) < 100:
            break
        else:
            parameters['cursor'] = json['cursor']
    print('done {i}'.format(i = index))
    steam_reviews.to_csv('Steam Reviews New.csv')
    time.sleep(60)
        
        
```
